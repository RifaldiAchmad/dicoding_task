### Pertanyaan 1: Kapan rentang waktu untuk jumlah pesanan menaik dan menurun?

**Insight:**

### Pertanyaan 2: Bagaimana distribusi review score paling kecil pada tiap produk?

---



```python
# Hitung jumlah review untuk review score 1 per kategori produk
score_counts = all_data[all_data['review_score'] == 1] \
    .groupby('product_category_name_english').size().reset_index(name='count')
```

```python
# Gabungkan dengan total review per kategori untuk menghitung persentase
total_counts = all_data.groupby('product_category_name_english').size().reset_index(name='total')
merged = pd.merge(total_counts, score_counts, on='product_category_name_english', how='left')
merged['count'] = merged['count'].fillna(0)
merged['percentage'] = (merged['count'] / merged['total']) * 100
```

```python
# Urutkan berdasarkan persentase tertinggi dan ambil top 10 kategori
top_10 = merged.sort_values(by='percentage', ascending=False).head(10)

# Plot bar chart untuk review score 1
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(top_10['product_category_name_english'], top_10['percentage'], color='royalblue')
ax.set_title(f"Review Score: 1", fontsize=14)
ax.set_ylabel('Persentase (%)', fontsize=12)
ax.tick_params(axis='x', rotation=90, labelsize=10)
ax.tick_params(axis='y', labelsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan nilai persentase di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Select the specified product categories
selected_categories = ['security_and_services', 'la_cuisine', 'fashion_male_clothing',
                       'home_comfort_2', 'diapers_and_hygience', 'office_furniture']

# Filter the DataFrame for the selected categories
filtered_data = all_data[all_data['product_category_name_english'].isin(selected_categories)]
category_counts = filtered_data['product_category_name_english'].value_counts()
category_counts_df = pd.DataFrame({'Category': category_counts.index, 'Count': category_counts.values})
category_counts_df
```

## Analisis Lanjutan (Opsional)

```python
# Mengelompokkan data dari DataFrame 'all_data' berdasarkan 'customer_id'
rfm = all_data.groupby("customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "nunique",
    "price": "sum"
})
rfm.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
```

```python
# Ubah kolom "max_order_timestamp" pada DataFrame rfm menjadi tipe date
rfm["max_order_timestamp"] = rfm["max_order_timestamp"].dt.date

# Mendapatkan tanggal pembelian terbaru (maksimum) dari seluruh data di all_data
recent_date = all_data["order_purchase_timestamp"].dt.date.max()
rfm["recency"] = rfm["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm.drop("max_order_timestamp", axis=1, inplace=True)
rfm.sort_values("monetary", ascending=False).reset_index(drop=True)
```

```python
# Menghitung peringkat (rank) untuk masing-masing metrik RFM:
rfm['r_rank'] = rfm['recency'].rank(ascending=False)
rfm['f_rank'] = rfm['frequency'].rank(ascending=True)
rfm['m_rank'] = rfm['monetary'].rank(ascending=True)
rfm
```

```python
# Normalisasi peringkat untuk setiap metrik RFM ke dalam skala 0-100
rfm['r_rank_norm'] = (rfm['r_rank']/rfm['r_rank'].max())*100
rfm['f_rank_norm'] = (rfm['f_rank']/rfm['f_rank'].max())*100
rfm['m_rank_norm'] = (rfm['m_rank']/rfm['m_rank'].max())*100

rfm.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)
rfm
```

```python
rfm['RFM_score'] = 0.15 * rfm['r_rank_norm'] + 0.28 * rfm['f_rank_norm'] + 0.57*rfm['m_rank_norm']
rfm['RFM_score'] *= 0.05
rfm = rfm.round(2)
rfm[['customer_id', 'RFM_score']].sort_values('RFM_score', ascending=False).reset_index(drop=True)
```

```python
rfm["customer_segment"] = np.where(
    rfm['RFM_score'] > 4.5, "Top", (np.where(
        rfm['RFM_score'] > 4, "High value",(np.where(
            rfm['RFM_score'] > 3, "Medium value", np.where(
                rfm['RFM_score'] > 1.6, 'Low value', 'lost'))))))

rfm[['customer_id', 'RFM_score', 'customer_segment']]
```

```python
# Menghitung jumlah pelanggan unik di setiap segmen pelanggan dan mengurutkannya
customer_segment_counts = (
    rfm.groupby("customer_segment", as_index=False)["customer_id"]
    .nunique()
    .sort_values("customer_id", ascending=False)
    .reset_index(drop=True)
)

# Mengonversi kolom 'customer_segment' menjadi tipe data kategori dengan urutan tertentu
customer_segment_counts["customer_segment"] = pd.Categorical(
    customer_segment_counts["customer_segment"],
    categories=["lost", "Low value", "Medium value", "High value", "Top"],
    ordered=True
)

# Menentukan ukuran figur
plt.figure(figsize=(10, 5))

# Menentukan warna untuk setiap segmen pelanggan
color_palette = ["#D3D3D3", "#72BCD4", "#72BCD4", "#D3D3D3", "#D3D3D3"]

# Membuat bar plot berdasarkan jumlah pelanggan di tiap segmen, diurutkan sesuai kategori
ax = sns.barplot(
    x="customer_id",
    y="customer_segment",
    data=customer_segment_counts.sort_values("customer_segment", ascending=False),
    palette=color_palette
)

# Menambahkan judul dan mengatur tampilan sumbu
plt.title("Jumlah Pelanggan Tiap Segmen", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis="y", labelsize=12)

# Menambahkan label jumlah pelanggan di setiap bar
label_offset = customer_segment_counts["customer_id"].max() * 0.01
for bar in ax.patches:
    customer_count = bar.get_width()  # Nilai jumlah pelanggan
    ax.text(
        customer_count + label_offset,
        bar.get_y() + bar.get_height() / 2,
        f"{int(customer_count)}",
        va="center"
    )
plt.show()
```
